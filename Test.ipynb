```
import tensorflow as tf
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.regularizers import l1, l2# Network and training parameters.
from io import BytesIO
import dill,base64,tempfile
#Loading Back
from joblib import load
from tensorflow.keras.models import model_from_json

import onnx
import os
#os.environ['TF_KERAS'] = '1'
import tf2onnx #keras2onnx,
import onnxruntime
import pickle

import google.protobuf.message
from typing import Union
```

```
def Base64Converter(ObjectFile):
    bytes_container = BytesIO()
    dill.dump(ObjectFile, bytes_container)
    bytes_container.seek(0)
    bytes_file = bytes_container.read()
    base64File = base64.b64encode(bytes_file)
    return base64File
def ObjectConverter(base64_File):
    loaded_binary = base64.b64decode(base64_File)
    loaded_object = tempfile.TemporaryFile()
    loaded_object.write(loaded_binary)
    loaded_object.seek(0)
    ObjectFile = load(loaded_object)
    loaded_object.close()
    return ObjectFile
```

```
EPOCHS = 5
BATCH_SIZE = 16
VERBOSE = 1
NB_CLASSES = 10  # number of outputs = number of digits
N_HIDDEN = 512
VALIDATION_SPLIT = 0.2  # how much TRAIN is reserved for VALIDATION
DROPOUT = 0.3
input_shape = (28, 28, 1)
mnist = keras.datasets.mnist
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
RESHAPED = 784
X_train = X_train.reshape(60000, RESHAPED)
X_test = X_test.reshape(10000, RESHAPED)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
# Normalize inputs to be within in [0, 1].
X_train /= 255
X_test /= 255
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
# One-hot representation of the labels.
Y_train = tf.keras.utils.to_categorical(Y_train, NB_CLASSES)
Y_test = tf.keras.utils.to_categorical(Y_test, NB_CLASSES)

model = tf.keras.models.Sequential()
model.add(keras.layers.Dense(units=64,name='layer_1',input_shape=(RESHAPED,),activation='relu'))
model.add(keras.layers.Dense(units=64,name='layer_2',input_shape=(RESHAPED,),activation='relu'))
model.add(keras.layers.Dense(units=64,name='layer_3',input_shape=(RESHAPED,),activation='relu'))
model.add(keras.layers.Dense(units=NB_CLASSES,name='layer_output',input_shape=(RESHAPED,),activation='softmax'))
model.summary()
model.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
```

```
#Load model Json


#0
my_model=model.to_json()
my_model_weight=model.get_weights()
base64KModelJson = Base64Converter(my_model)
base64KModelJsonWeights = Base64Converter(model.get_weights())

modeljson = ObjectConverter(base64KModelJson)
modelweights = ObjectConverter(base64KModelJsonWeights)
loaded_model = model_from_json(modeljson)
loaded_model.set_weights(modelweights)
loaded_model.summary()

loaded_model.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
loaded_model.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
#1
print("step 1")
loaded_model1 = model_from_json(ObjectConverter(Base64Converter(loaded_model.to_json())))
loaded_model1.set_weights(ObjectConverter(Base64Converter(loaded_model.get_weights())))
loaded_model1.summary()
loaded_model1.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
loaded_model1.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
#2
print("step 2")
loaded_model2 = model_from_json(ObjectConverter(Base64Converter(loaded_model1.to_json())))
loaded_model2.set_weights(ObjectConverter(Base64Converter(loaded_model1.get_weights())))
loaded_model2.summary()
loaded_model2.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
loaded_model2.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
#3
print("step 3")
loaded_model3 = model_from_json(ObjectConverter(Base64Converter(loaded_model2.to_json())))
loaded_model3.set_weights(ObjectConverter(Base64Converter(loaded_model2.get_weights())))
loaded_model3.summary()
loaded_model3.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
loaded_model3.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
#4
print("step 4")

loaded_model4 = model_from_json(ObjectConverter(Base64Converter(loaded_model3.to_json())))
loaded_model4.set_weights(ObjectConverter(Base64Converter(loaded_model3.get_weights())))
loaded_model4.summary()
loaded_model4.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
loaded_model4.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
```

```
def _serialize(proto: Union[bytes, google.protobuf.message.Message]) -> bytes:
    if isinstance(proto, bytes):
        return proto
    elif hasattr(proto, 'SerializeToString') and callable(proto.SerializeToString):
        result = proto.SerializeToString()
        return result
    else:
        raise TypeError('No SerializeToString method is detected. '
                         'neither proto is a str.\ntype is {}'.format(type(proto)))
```

```
onnx_model, _ = tf2onnx.convert.from_keras(model)
onx_dum=_serialize(onnx_model)

session_b = onnxruntime.InferenceSession(onx_dum)
```

```
#Merge bytes
leng_header=4096
leng_model=len(base64KModelJson)
leng_weight=len(base64KModelJsonWeights)
leng_onnxdum=len(onx_dum)
```

```
buffer_all=[leng_header+leng_model+leng_weight+leng_onnxdum]
buffer_all[leng_header:leng_header+leng_model-1]=base64KModelJson
buffer_all[leng_header+leng_model:leng_header+leng_model+leng_weight-1]=base64KModelJsonWeights
buffer_all[leng_header+leng_model+leng_weight:leng_header+leng_model+leng_weight+leng_onnxdum-1]=onx_dum

buffer_all_in_one=base64KModelJson+base64KModelJsonWeights+onx_dum
print(len(buffer_all_in_one))

#Decoder
model_re=buffer_all_in_one[0:leng_model]

model_w=buffer_all_in_one[leng_model:leng_model+leng_weight]

loaded_model_recontruct = model_from_json(ObjectConverter(model_re))
loaded_model_recontruct.set_weights(ObjectConverter(model_w))
loaded_model_recontruct.summary()
loaded_model_recontruct.compile(optimizer='SGD',loss='categorical_crossentropy',metrics=['accuracy'])
loaded_model_recontruct.fit(X_train, Y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
```
